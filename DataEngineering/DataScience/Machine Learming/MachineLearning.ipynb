```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.cluster import KMeans
import numpy as np


# Correctly define categorical_columns
categorical_columns = ['course_name', 'outcome_skill', 'role_outcome', 
                       'current_role', 'skills', 'feedback_category', 
                       'yearofexperience_category']

# Preprocess: Convert columns to string and strip whitespace
for column in categorical_columns:
    df[column] = df[column].astype(str).str.strip()  # Convert to string and strip whitespace

# Use OneHotEncoder with handle_unknown='ignore'
encoder = OneHotEncoder(drop='first', handle_unknown='ignore')
encoded_features = encoder.fit_transform(df[categorical_columns]).toarray()

# Fit K-Means Clustering
n_clusters = 9  # Choose the number of clusters based on your data
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(encoded_features)

# Add cluster labels to the original DataFrame
df['cluster'] = kmeans.labels_

# Function to recommend a course based on a new employee's data
def recommend_courses_based_on_role(new_employee_data, encoder, kmeans, df):
    # Clean and prepare the new employee data
    if len(new_employee_data) != len(categorical_columns):
        raise ValueError("New employee data must match the number of categorical columns.")

    # Convert the new employee data to a DataFrame
    new_employee_df = pd.DataFrame([new_employee_data], columns=categorical_columns)

    # Ensure all values are in the expected format (string and stripped)
    for column in categorical_columns:
        new_employee_df[column] = new_employee_df[column].astype(str).str.strip()

    # Get the desired role outcome and current skills
    desired_role_outcome = new_employee_df['role_outcome'].values[0]
    current_role = new_employee_df['current_role'].values[0]
    current_skills = set(new_employee_df['skills'].values[0].split(','))

    # Filter courses based on the desired role outcome and relevant skills
    relevant_courses = df[df['role_outcome'] == desired_role_outcome]

    # Further filter by checking if the skills are related to the courses
    recommended_courses = []
    for _, course in relevant_courses.iterrows():
        course_skills = set(course['outcome_skill'].split(','))
        if not current_skills.intersection(course_skills):  # Recommend if there is no skill match
            recommended_courses.append(course['course_name'])

    return list(set(recommended_courses))  # Return unique course names

# New employee data (example)
new_employee_data = ['Building Data Pipelines', 'Data Modeling', 
                     'junior data engineer', 'junior software engineer', 
                     'Microservices, Sql', 'Neutral', '5-10']

# Get recommended courses
recommended_courses = recommend_courses_based_on_role(new_employee_data, encoder, kmeans, df)
print(f'Recommended Courses: {recommended_courses}')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Initialize a list to store user input and output data
user_data_storage = []

# Function for hybrid course recommendation
def hybrid_recommendation(employee_id, desired_role_outcome, current_skills, completed_courses, years_of_experience, current_role, df):
    current_skills_set = set(current_skills.split(','))
    completed_courses_set = {course.strip() for course in completed_courses.split(',')}  # Handle multiple completed courses

    # Filter courses based on the desired role outcome
    relevant_courses = df[df['role_outcome'] == desired_role_outcome].copy()

    # Create a TF-IDF vectorizer for the outcome skills
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(relevant_courses['outcome_skill'])

    # Prepare a string combining current skills and completed courses
    combined_skills = ' '.join(current_skills_set) + ' ' + ' '.join(completed_courses_set)
    employee_vector = vectorizer.transform([combined_skills])

    # Compute cosine similarity between the employee skills and course skills
    cosine_similarities = cosine_similarity(employee_vector, tfidf_matrix)

    # Add similarity scores to the relevant courses DataFrame
    relevant_courses['similarity'] = cosine_similarities.flatten()

    # Sort courses by similarity score and return the top recommendations
    recommended_courses = relevant_courses.sort_values(by='similarity', ascending=False)['course_name'].unique().tolist()

    # Filter out courses that the user has already completed
    recommended_courses = [course for course in recommended_courses if course not in completed_courses_set]

    # Suggest skills needed for the desired role outcome
    suggested_skills = set()
    for _, course in relevant_courses.iterrows():
        course_skills = set(course['outcome_skill'].split(','))
        suggested_skills.update(course_skills)

    # Remove skills that the user already has
    suggested_skills = [skill for skill in suggested_skills if skill not in current_skills_set]

    # Ensure suggested skills are unique
    suggested_skills = list(set(suggested_skills))

    # Store the user input and output in the data storage
    user_data_storage.append({
        'employee_id': employee_id,
        'input_role_outcome': desired_role_outcome,
        'input_current_skills': list(current_skills_set),
        'completed_courses': list(completed_courses_set),
        'years_of_experience': years_of_experience,
        'current_role': current_role,
        'recommended_courses': recommended_courses,
        'suggested_skills': suggested_skills
    })

    return recommended_courses, suggested_skills

# Collect user input
employee_id = input("Please enter your Employee ID: ")
desired_role_outcome = input("Please enter your desired role outcome: ")
current_skills = input("Please enter your current skills (comma-separated): ")
completed_courses = input("Please enter the names of any completed courses (comma-separated): ")
years_of_experience = input("Please enter your years of experience: ")
current_role = input("Please enter your current role: ")

# Assuming df is already defined and populated with course data
recommended_courses, suggested_skills = hybrid_recommendation(
    employee_id,
    desired_role_outcome,
    current_skills,
    completed_courses,
    years_of_experience,
    current_role,
    df
)

# Print the results
print(f"Recommended Courses: {set(recommended_courses)}")  # Proper set formatting
print(f"Suggested Skills: {suggested_skills}")

# Convert user data storage to a DataFrame
user_data_df = pd.DataFrame(user_data_storage)

# Append to CSV file
user_data_df.to_csv('user_data_storage.csv', mode='a', header=not pd.io.common.file_exists('user_data_storage.csv'), index=False)

# To confirm the data was saved
print("\nData saved to user_data_storage.csv")
```
